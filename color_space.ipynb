# ***Original Image***

```python
import cv2
from google.colab.patches import cv2_imshow

imageb1 = cv2.imread('/content/001_03.jpg')
imageb2 = cv2.imread('/content/010_03.jpg')

imagem = cv2.imread('/content/001_03-vs-010_03.png')


imgb1 = cv2.resize(imageb1, (300, 300))
imgb2 = cv2.resize(imageb2, (300, 300))
imgm = cv2.resize(imagem, (300, 300))

# cv2_imshow(imgb1)
# cv2_imshow(imgb2)
# cv2_imshow(imgm)
```

```python
import cv2
import matplotlib.pyplot as plt

def display_images_in_line(image1, image2, image3, common_title):
    # Convert BGR images to RGB for compatibility with Matplotlib
    image1_rgb = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
    image2_rgb = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
    image3_rgb = cv2.cvtColor(image3, cv2.COLOR_BGR2RGB)

    # Create a figure with three subplots
    fig, axs = plt.subplots(1, 3, figsize=(12, 4))
    fig.suptitle(common_title, fontsize=20, fontweight='bold')


    # Display the images in subplots
    axs[0].imshow(image1_rgb)
    axs[0].axis('off')
    axs[0].set_title('Subject 1')

    axs[1].imshow(image3_rgb)
    axs[1].axis('off')
    axs[1].set_title('Subject 1 + Subject 2')

    axs[2].imshow(image2_rgb)
    axs[2].axis('off')
    axs[2].set_title('Subject 2')

    # Adjust the spacing between subplots
    plt.tight_layout()

    # Show the plot
    plt.show()
```

```python
display_images_in_line(imgb1, imgb2, imgm, "Original Images")
```

# ***RGB to YCbCr***

```python
ycbcrb1 = cv2.cvtColor(imgb1, cv2.COLOR_BGR2YCrCb)
ycbcrb2 = cv2.cvtColor(imgb2, cv2.COLOR_BGR2YCrCb)
ycbcrm  = cv2.cvtColor(imgm, cv2.COLOR_BGR2YCrCb)
# cv2_imshow(ycbcrb1)
# cv2_imshow(ycbcrm)
# cv2_imshow(ycbcrb2)
```

```python
display_images_in_line(ycbcrb1, ycbcrb2, ycbcrm, "Applied YCbCr")
```

# ***RGB to HSV***

```python
hsvb1 = cv2.cvtColor(imgb1, cv2.COLOR_BGR2HSV)
hsvb2 = cv2.cvtColor(imgb2, cv2.COLOR_BGR2HSV)
hsvm  = cv2.cvtColor(imgm, cv2.COLOR_BGR2HSV)

# cv2_imshow(hsvb1)
# cv2_imshow(hsvm)
# cv2_imshow(hsvb2)
```

```python
display_images_in_line(hsvb1, hsvb2, hsvm, "Applied HSV")
```

## ***Split YCbCr in Y, Cb, Cr***

```python
import cv2
import matplotlib.pyplot as plt

def display_images_in_line4(image1, image2, image3, image4, common_title):
    # Convert BGR images to RGB for compatibility with Matplotlib
    image1_rgb = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
    image2_rgb = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
    image3_rgb = cv2.cvtColor(image3, cv2.COLOR_BGR2RGB)
    image4_rgb = cv2.cvtColor(image4, cv2.COLOR_BGR2RGB)

    # Create a figure with four subplots
    fig, axs = plt.subplots(1, 4, figsize=(16, 4))
    fig.suptitle(common_title, fontsize=20, fontweight='bold')

    # Display the images in subplots
    axs[0].imshow(image1_rgb)
    axs[0].axis('off')
    axs[0].set_title('Original')

    axs[1].imshow(image2_rgb)
    axs[1].axis('off')
    axs[1].set_title('Y-Luminance, Iy')

    axs[2].imshow(image3_rgb)
    axs[2].axis('off')
    axs[2].set_title('Cb-Blue Chromaticity, Icb')

    axs[3].imshow(image4_rgb)
    axs[3].axis('off')
    axs[3].set_title('Cr-Red Chromaticity, Icr')

    # Adjust the spacing between subplots
    plt.tight_layout()

    # Show the plot
    plt.show()
```

```python
yb1, cbb1, crb1 = cv2.split(ycbcrb1)
display_images_in_line4(ycbcrb1, yb1, cbb1, crb1, "Splitted YCbCr for subject 1")
```

```python
yb2, cbb2, crb2 = cv2.split(ycbcrb2)
display_images_in_line4(ycbcrb2, yb2, cbb2, crb2, "Splitted YCbCr for subject 2")
```

```python
ybm, cbbm, crbm = cv2.split(ycbcrm)
display_images_in_line4(ycbcrm, ybm, cbbm, crbm, "Splitted YCbCr for subject 1 + subject 2")
```

## ***Split HSV in H, S, V***

```python
import cv2
import matplotlib.pyplot as plt

def display_images_in_line4(image1, image2, image3, image4, common_title):
    # Convert BGR images to RGB for compatibility with Matplotlib
    image1_rgb = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
    image2_rgb = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
    image3_rgb = cv2.cvtColor(image3, cv2.COLOR_BGR2RGB)
    image4_rgb = cv2.cvtColor(image4, cv2.COLOR_BGR2RGB)

    # Create a figure with four subplots
    fig, axs = plt.subplots(1, 4, figsize=(16, 4))
    fig.suptitle(common_title, fontsize=20, fontweight='bold')

    # Display the images in subplots
    axs[0].imshow(image1_rgb)
    axs[0].axis('off')
    axs[0].set_title('Original')

    axs[1].imshow(image2_rgb)
    axs[1].axis('off')
    axs[1].set_title('H-Hue, Ih')

    axs[2].imshow(image3_rgb)
    axs[2].axis('off')
    axs[2].set_title('S-Saturation, Is')

    axs[3].imshow(image4_rgb)
    axs[3].axis('off')
    axs[3].set_title('V-Value, Iv')

    # Adjust the spacing between subplots
    plt.tight_layout()

    # Show the plot
    plt.show()
```

```python
hb1, sb1, bb1 = cv2.split(hsvb1)
display_images_in_line4(hsvb1, hb1, sb1, bb1, "Splitted HSV for subject 1")
```

```python
hb2, sb2, bb2 = cv2.split(hsvb2)
display_images_in_line4(hsvb2, hb2, sb2, bb2, "Splitted HSV for subject 2")
```

```python
hm, sm, bm = cv2.split(hsvm)
display_images_in_line4(hsvm, hm, sm, bm, "Splitted HSV for subject 1 + subject 2")
```
